```python
def hnormal(inicioA, terminoA, inicioJ, terminoJ, dia):
    hor=[timedelta(),timedelta()]
    faixa_normal=dia+timedelta(hours=22)
    if(inicioA>=inicioJ)&(terminoA<=terminoJ):
        if(terminoA<=faixa_normal):
            hor[0]=hor[0]+ (terminoA - inicioA)
        elif(terminoA>faixa_normal)&(inicioA<faixa_normal):
            hor[0]=hor[0]+(faixa_normal - inicioA)
            hor[1]=hor[1]+(terminoA-faixa_normal)
        else:
            hor[1]=(terminoA - inicioA)
    return hor
```

```python
def addnot(inicioA, terminoA, inicioJ, terminoJ, dia):
    addnot=timedelta()
    faixa_normal=dia+timedelta(hours=22)
    if(inicioA>=inicioJ)&(terminoA<=terminoJ):
        if(terminoA>faixa_normal):
            addnot=addnot+ (terminoA - inicioA)
        else:
            addnot=addnot+(faixa_normal - inicioA)
    return addnot
```
